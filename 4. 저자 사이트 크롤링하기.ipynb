크롤링할 사이트의 주소

https://pjt3591oo.github.io

1. 우클릭 했을 때 페이지 소스 보기만 있는가 : YES
2. 원하는 문자열이 소스에 있는가 : YES

```python
import bs4
import requests
import time
```

```python
# 첫 페이지의 주소
site = 'https://pjt3591oo.github.io'
# 수집된 데이터를 담을 리스트
data_list = []

while True :
    time.sleep(1)
    # div 태그 리스트를 가져온다.
    div_list, next_target = getDivList(site)
    # 태그의 개수만큼 반복한다.
    for div_tag in div_list :
        # div 태그 내의 데이터를 추출한다.
        tuple1 = getData(div_tag)
        # print(tuple1)
        # 데이터들의 이름 튜플
        tuple2 = ('href', 'a_text', 'h4_text', 'date_text', 'author_text')
        # 딕셔너리로 만든다.
        zip1 = zip(tuple2, tuple1)
        # print(tuple(zip1))
        data_dic = dict(zip1)
        # print(data_dic)
        data_list.append(data_dic)
    
    print(next_target)
    # 다음 페이지로 이동한다.
    if next_target != None :
        site = f'https://pjt3591oo.github.io/{next_target}' 
    else :
        break
    
    
# 출력한다.
printList(data_list)
```

```python
# 주소를 받아 요청하고 class가 p인 div 태그 
# 리스트를 반환한다.
def getDivList(site) :
    # 요청
    response = requests.get(site)
    # 응답 결과를 가져와 태그 객체를 만든다.
    soup = bs4.BeautifulSoup(response.text)
    # class가 p인 div 태그 객체 리스트를 추출한다.
    div_list = soup.find_all('div', class_='p')
    
    # 다음 페이지가 있는지 확인한다.
    # Next 문자열 값을 가지고 있는 태그를 가져온다.
    next_tag = soup.find(any, text='Next')
    next_target = None
    if next_tag.name == 'a' :
        next_target = next_tag.attrs['href']
    
    # 반환한다.
    return div_list, next_target
```

```python
# 실제 데이터를 추출해서 반환하는 함수
def getData(div_tag) :
    # a 태그를 가져온다.
    h3_tag = div_tag.find('h3')
    a_tag = h3_tag.find('a')
    href = a_tag.attrs['href']
    a_text = a_tag.text.strip()
    
    # h4
    h4_tag = div_tag.find('h4')
    h4_text = h4_tag.text.strip()
    
    # date, author
    p_tag = div_tag.find('p', class_='author')
    span_tag = p_tag.find('span', class_='date')
    temp_str = span_tag.text.strip()
    temp_list = temp_str.split('|')
    date_str = temp_list[0].strip()
    author_str = temp_list[1].strip()
    
    return href, a_text, h4_text, date_str, author_str
```

```python
# 리스트에 담긴 데이터를 출력한다.
def printList(data_list) :
    for data_dic in data_list :
        for key, value in data_dic.items() :
            print(f'{key} : {value}')
        print("----------------------")
```
