# RSS 문서

국제 표준으로 양식이 정해져 있는 문서

&lt;rss version=2.0&gt; 으로 시작하면 RSS문서다

rss 문서의 구조와 태그는 정해져 있는 태그를 사용하고 있고 항목 하나의 데이터를 가지고 있는 item 태그 내부는 자유롭게 구성되어 있다. 따라서 item 태그 내부만 살펴보면 쉽게 데이터를 가져올 수 있다.

```python
# bs4 사용
import requests
import bs4

site = 'http://aladin.co.kr/rss/special_new/351'

response = requests.get(site)
soup = bs4.BeautifulSoup(response.text, 'lxml-xml')

# 전체 타이틀 추출
title = soup.select('channel > title')[0].text
print(f'title : {title}')
# 전체 링크 추출
link = soup.select('channel > link')[0].text
print(f'link : {link}')

# item 태그들을 가져온다.
item_list = soup.select('item')
for item_tag in item_list : 
    item_title = item_tag.select('title')[0].text
    item_link = item_tag.select('link')[0].text
    print(f'item_title : {item_title}')
    print(f'item_link : {item_link}')
    print('------------------------------')
```

```python
# feedparser 사용
# pip install feedparser
import feedparser

site = 'http://aladin.co.kr/rss/special_new/351'

# 파싱한다.
result = feedparser.parse(site)

#  출력
print(f'title : {result.feed.title}')
print(f'link : {result.feed.link}')

# item 태그 내부는 리스트로 만들어지고 entries라는
# 곳에 담겨있다.
for entry in result.entries :
    print(f'item_title : {entry.title}')
    print(f'item_link : {entry.link}')
    print('---------------------------------')
```
